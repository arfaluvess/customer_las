# Second way to access directly using sheet

```python
# Import libraries
from io import BytesIO
from google.auth import default as get_google_default
from google.colab import auth
from googleapiclient.discovery import build
from gspread import authorize as gspread_auth
from gspread_dataframe import get_as_dataframe
from pandas import DataFrame, Series, read_excel
from re import compile as re
from tqdm.auto import tqdm

auth.authenticate_user()
gcreds, _ = get_google_default()
gc = gspread_auth(gcreds)
drive = build('drive', 'v3', credentials=gcreds)
drive_home = '/content'
drive_files = drive.files()

# Get the sheet to read as a dataframe
def gspread_to_df(book_id:str, sheet_name:str) -> DataFrame:
    _gbook = gc.open_by_key(book_id)
    _gsheet = get_as_dataframe(_gbook.worksheet(sheet_name), evaluate_formulas=True)
    _gsheet.dropna(axis='index', how='all', inplace=True)
    _gsheet.dropna(axis='columns', how='all', inplace=True)
    return _gsheet
```

```python
# Convert to a formatted MY phone number
import phonenumbers

def standardize_phone_number(phone_number):
  try:
    parsed_number = phonenumbers.parse(phone_number, "MY")
    if phonenumbers.is_possible_number(parsed_number) and phonenumbers.is_valid_number(parsed_number):
      return phonenumbers.format_number(parsed_number, phonenumbers.PhoneNumberFormat.E164)
    else:
      print(f"Invalid phone number: {phone_number}")
      return None 
  except phonenumbers.phonenumberutil.NumberParseException as e:
    print(f"Error while parsing phone number: {phone_number}")
    return None

cus_df_29['phone_number'] = cus_df_29['phone_number'].apply(standardize_phone_number)
```

# Save to new sheet 

- Save new dataframe inside existing sheet 
- Formatting the data

```python
import gspread
from gspread_dataframe import set_with_dataframe
from googleapiclient.discovery import build 


try: 
  # Authenticate google sheets API credential
  gc = gspread.authorize(creds)
  
  # Open google sheets by its title
  sh = gc.open('PRD Payment Report by Month [H]')

  # Add a new sheet to the existing Google Sheet
  worksheet = sh.add_worksheet(title="umrah_payment_due", 
                              rows=column_df.shape[0], 
                              cols=column_df.shape[1])

  # # Save the DataFrame to the new sheet
  set_with_dataframe(worksheet, column_df)
  print('Google sheet opened successfully')

except gspread.exceptions.APIError as e:
  if e.response.status_code == 400:
    if 'A sheet with the name' in e.response.text:
      worksheet = sh.worksheet('umrah_payment_due')
      worksheet.clear()
      print('Sheet already exist, clearing data')
      # Save the DataFrame to the new sheet
      set_with_dataframe(worksheet, column_df)
      print('Sheet updated with new data')

    else:
      print('An error occured: ', e)
```
